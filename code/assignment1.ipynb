```python
# Task 1.1: Data Retrieving
# Importing pandas, numpy and pyplot packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing StarWars.csv dataset and assigning it to a variable using read_csv function
starwars_file = 'StarWars.csv' 
starwars_unc = pd.read_csv(starwars_file, sep=',', decimal = '.')

# the top row contains the response options from the survey 
starwars_unc.head(3)
```

```python
# 3rd - 9th Column
c3 = 'Do you consider yourself to be a fan of the Star Wars film franchise?'
c9 = 'starwars_6'
starwars.loc[:, c3:c9] = starwars.loc[:, c3:c9].replace(np.NaN, 'No')
```

```python
# 4th - 9th Column
moviemap2yes = {'Star Wars: Episode I  The Phantom Menace' : 'Yes',
                'Star Wars: Episode II  Attack of the Clones' : 'Yes',
                'Star Wars: Episode III  Revenge of the Sith' : 'Yes',
                'Star Wars: Episode IV  A New Hope' : 'Yes',
                'Star Wars: Episode V The Empire Strikes Back' : 'Yes',
                'Star Wars: Episode VI Return of the Jedi' : 'Yes'}
starwars.replace(moviemap2yes, inplace = True)
```

```python
# 10th - 15th Column. c10 = 'rank_1', c15 = 'rank_6'
starwars.loc[:, 'rank_1':'rank_6'] = starwars.loc[:, 'rank_1':'rank_6'].fillna(0)
```

```python
# 16th - 29th Column [No, NaN]
c2 = 'Have you seen any of the 6 films in the Star Wars franchise?'
for i in range(15,29):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]]
    starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'Unfamiliar (N/A)')
```

```python
# 16th - 29th Column [Yes, NaN]
for i in range(15,29):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]]
    starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'Neither favorably nor unfavorably (neutral)') 
```

```python
# 30th Column [No, NaN]
c30 = 'Which character shot first?'
rep = starwars.loc[(starwars[c2] == 'No') & (starwars[c30].isna()), c30]
starwars.loc[(starwars[c2] == 'No') & (starwars[c30].isna()), c30] = rep.replace(np.NaN, "I don't understand this question")
```

```python
# 30th Column [Yes, NaN]
rep = starwars.loc[(starwars[c2] == 'Yes') & (starwars[c30].isna()), c30]
starwars.loc[(starwars[c2] == 'Yes') & (starwars[c30].isna()), c30] = rep.replace(np.NaN, 'Han')
```

```python
# 31st to 33rd Columns [No, NaN]
for i in range(30,33):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]]
    starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'No') 
```

```python
# 31st to 33rd Columns [Yes, NaN]
for i in range(30,33):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]]
    if i == 30:      
        starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'No') 
    if i == 31:
        # Note this condition should be executed only after replacing[No, NaN] values to No. 
        # As fillna needs to select whole column, doesmt work on specified condition
        starwars.loc[: , starwars.columns[i]].fillna( method ='ffill', inplace = True)
    if i == 32:      
        starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'No')
```

```python
# 34th - 35th Column
starwars.loc[: , starwars.columns[33]].fillna( method ='ffill', inplace = True)
starwars.loc[: , starwars.columns[34]].fillna( method ='ffill', inplace = True)
# 37th Column
starwars.loc[: , starwars.columns[36]].fillna('Some college or Associate degree', inplace = True) 
# 38th Colum
starwars.loc[: , starwars.columns[37]].fillna( method ='ffill', inplace = True)
```

```python
# Task 2.1: Explore a survey question
avg = starwars[starwars.columns[9:15]].sum()
index = starwars.columns[9:15]
plt.bar(index, avg)
plt.xlabel('Names of films')
plt.ylabel('Total ratings')
plt.title('Total ratings of all Star Wars films')
plt.show()
# Highest means Lowest 
```

```python
# Task 2.2: Relationships between columns
grouped = starwars.groupby(['Household Income','Gender'])['Household Income'].count()
myorder = ['$0 - $24,999', '$25,000 - $49,999', '$50,000 - $99,999', '$100,000 - $149,999', '$150,000+']
Ordered = grouped.reindex(myorder, level = 0)
Ordered.unstack().plot.bar()
```

```python
Ordered.unstack().plot.pie(subplots=True, figsize=(15, 10), autopct='%.2f%%')
```

```python
Ordered.unstack().plot(kind = 'density')
```

```python
p = starwars.groupby(['Household Income','Gender'])['Household Income'].count().unstack()
mdict = [i for i in p['Male']] 
fdict = [i for i in p['Female']]
```

```python
myorder = ['$0 - $24,999', '$25,000 - $49,999', '$50,000 - $99,999', '$100,000 - $149,999', '$150,000+']
plt.scatter(x = myorder, y = mdict, color='red')
plt.scatter(x = myorder, y = fdict, color='green')
plt.xticks(rotation=90)
plt.show()
```

```python
# Task 2.3: Explore a specific relationship. Gender/Yoda
mylist = ['Gender', 'Age', 'Household Income', 'Education', 'Location (Census Region)']
myorder = ['Very favorably', 'Somewhat favorably', 'Neither favorably nor unfavorably (neutral)', 'Somewhat unfavorably', 'Very unfavorably', 'Unfamiliar (N/A)']
for j in mylist:
    grouped = starwars.groupby(['Luke Skywalker', j])[j].count()
    Ordered = grouped.reindex(myorder, level = 0)
    Ordered.unstack().plot.bar(figsize = (8,5))
    plt.xlabel("Luke Skywalker")
```
