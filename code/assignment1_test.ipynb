```python
# 16th - 29th Column [No, NaN]
c2 = 'Have you seen any of the 6 films in the Star Wars franchise?'
for i in range(15,29):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]]
    starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'Unfamiliar (N/A)')
```

```python
# 16th - 29th Column [Yes, NaN]
for i in range(15,29):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]]
    starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'Neither favorably nor unfavorably (neutral)') 
```

```python
# 30th Column [No, NaN]
c30 = 'Which character shot first?'
rep = starwars.loc[(starwars[c2] == 'No') & (starwars[c30].isna()), c30]
starwars.loc[(starwars[c2] == 'No') & (starwars[c30].isna()), c30] = rep.replace(np.NaN, "I don't understand this question")
```

```python
# 30th Column [Yes, NaN]
rep = starwars.loc[(starwars[c2] == 'Yes') & (starwars[c30].isna()), c30]
starwars.loc[(starwars[c2] == 'Yes') & (starwars[c30].isna()), c30] = rep.replace(np.NaN, 'Han')
```

```python
# 31st to 33rd Columns [No, NaN]
for i in range(30,33):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]]
    starwars.loc[(starwars[c2] == 'No') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'No') 
```

```python
# 31st to 33rd Columns [Yes, NaN]
for i in range(30,33):
    col = starwars.columns[i]
    switch = starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]]
    if i == 30:      
        starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'No') 
    if i == 31:
        # Note this condition should be executed only after replacing[No, NaN] values to No. 
        # As fillna needs to select whole column, doesmt work on specified condition
        starwars.loc[: , starwars.columns[i]].fillna( method ='ffill', inplace = True)
    if i == 32:      
        starwars.loc[(starwars[c2] == 'Yes') & (starwars[col].isna()), starwars.columns[i]] = switch.replace(np.NaN, 'No')
```
